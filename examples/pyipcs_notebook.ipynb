# pyIPCS Example Jupyter Notebook

## Insert Dump Dataset Name

```python
from pyipcs import IpcsSession, Subcmd

# Insert z/OS Dump Dataset Name

dsname = ""

print(dsname)
```

## Configure pyIPCS Session

***Note**: Do not overwrite `session`(IpcsSession) variable while session is active/open*

```python
# You May Need to Configure Your pyIPCS Session:

# Configure your Allocations
#   `IpcsSession.aloc`

# Configure Your DDIR Defaults or DDIR Creation Settings
#   `IpcsSession.ddir`

session = IpcsSession()
```

## Open pyIPCS Session 

***Note:** Make sure to run `session.close()` at the end of this notebook when you are done*

```python
session.open()

print("pyIPCS Session Opened")
```

## Initialize Dump

Will initialize dump under temporary DDIR that will be deleted on session close

```python
# Create Dump object `dump`
# Initializes z/OS dump and stores general info

print("Begin Dump Initialization")

dump = session.init_dump(dsname)

print(dump)
```

## Get Info About the Dump

```python
print(f"Dump Title: {dump.header['title']}")

print(f"Version: {dump.header['version']}")

print(f"Release: {dump.header['release']}")
```

## Run Subcommand `STATUS REGISTERS`

```python
subcmd = Subcmd(session, "STATUS REGISTERS")

print(subcmd)
```

## Get Data From `STATUS REGISTERS`

### First 20 Characters

```python
print(repr(subcmd[:20]))
```

### Full Output

```python
print(subcmd.output)
```

## Close pyIPCS Session

```python
session.close()

print("pyIPCS Session Closed")
```
